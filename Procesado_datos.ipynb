## Tratamiento datos info_da

#### Limpiar nas:

``````python
lista_nas = []

for i in list(df_da.columns):
    lista_nas.append(df_da[i].isnull().values.any())
    

lista_col_nas = []

for i in range(len(lista_nas)):
    if lista_nas[i] == True:
        lista_col_nas.append(df_da.columns[i])
``````

``````python
lista_col_nas
``````

``````python
df_da['Puesto ofertado'].fillna('no_data', inplace=True)
``````

``````python
df_da['Ubicación'].fillna('no_data', inplace=True)
``````

``````python
df_da['Días activos'].fillna('no_data', inplace=True)
``````

``````python
df_da['Nº Solicitudes'].fillna(0, inplace=True)
``````

``````python
df_da['Nivel experiencia'].fillna('no_data', inplace=True)
``````

## Limpiar datos columnas:

#### Ubicación:

``````python
df_da['Ubicación'] = df_da['Ubicación'].str.split(' ').str[0]
``````

``````python
df_da['Ubicación'] = df_da['Ubicación'].str.replace(',', '')
``````

#### Días activos:

``````python
df_da['Días activos'] = df_da['Días activos'].str.replace('Publicado hace hace', '')
``````

#### Solicitudes:

``````python
lista_rem_so = [' visualizaciones', ' visualización', '.']

for i in lista_rem_so:
    df_da['Nº Solicitudes'] = df_da['Nº Solicitudes'].str.replace(i, '')
``````

``````python
df_da['Nº Solicitudes'] = df_da['Nº Solicitudes'].astype(int)
``````

``````python
df_da['Nº Solicitudes'].head()
``````

#### Nivel experiencia:

``````python
mask_nexp = df_da['Nivel experiencia'].str.contains('Sin experiencia|Algo de responsabilidad|Intermedio|Prácticas')
``````

``````python
mask_nexp
``````

``````python
df_da.loc[-mask_nexp,'Nivel experiencia'] = 'no_data'
``````

#### Tipo jornada

``````python
mask_tjor = df_da['Tipo jornada'].str.contains('Internet')
``````

``````python
df_da.loc[-mask_tjor,'Tipo jornada'] = 'no_data'
``````

#### Guardado del fichero limpio

``````python
df_da.to_csv('info_data_analyst_clean.csv')
``````
