```python
df_da_or = pd.read_csv('info_data_analyst_clean.csv')
df_ds_or = pd.read_csv('info_data_scientist_clean.csv')
```

```python
df_da = df_da_or.copy()
df_ds = df_ds_or.copy()
```

```python
df_da.drop(columns = 'Unnamed: 0', inplace=True)
df_ds.drop(columns = 'Unnamed: 0', inplace=True)
```

```python
df_da.head(1)
```

### Concatenar los dos DF creando columna para agrupar por tipo.

```python
df_da['Tipo puesto'] = pd.Series('')
df_ds['Tipo puesto'] = pd.Series('')
```

```python
df_da['Tipo puesto'] = 'Data Analyst'
df_ds['Tipo puesto'] = 'Data Scientist'
```

```python
df = pd.concat([df_da, df_ds])
```

```python
df = df.reset_index()
```

```python
df.drop(columns = 'index', inplace=True)
```

```python
puesto_da = pd.Series(' '.join(df_da['Puesto ofertado']).lower().split()).value_counts()[:30]
```

```python
rem_word_puesto = ['-', 'and', 'analyst,', 'de', ',', '&', '–', 'provided)', '/', 'it']

for i in rem_word_puesto:
    if i in puesto_da:
        puesto_da = puesto_da.drop(i)
```

```python
puesto_da = puesto_da.drop(puesto_da.index[6])
```

```python
puesto_da = puesto_da[:12]
```

```python
puesto_da
```

```python
puesto_ds = pd.Series(' '.join(df_ds['Puesto ofertado']).lower().split()).value_counts()[:30]
```

```python
puesto_ds[0:20]
```

```python
rem_word_puesto = ['-', 'and', '&', '–']

for i in rem_word_puesto:
    if i in puesto_ds:
        puesto_ds = puesto_ds.drop(i)
```

```python
puesto_ds = puesto_ds[:12]
```

```python
puesto_ds.index
```

```python
# create trace1 
trace1 = go.Bar(
    x = puesto_da.index,
    y = puesto_da,
    name = 'Palabras puesto DA',
    marker = dict(color = 'rgba(255, 174, 255, 0.5)',
                 line = dict(color = 'rgb(0, 0, 0)', width = 1.5))
)

# Podemos ponerle el color de la línea de la barra.

trace2 = go.Bar(
    x = puesto_ds.index,
    y = puesto_ds.index,
    name = 'Palabras puesto DS',
    marker = dict(color = 'rgba(255, 255, 128, 0.5)',
                 line = dict(color = 'rgb(0, 0, 0)', width = 1.5))
)

data = [trace1, trace2]
layout = go.Layout(barmode = "group")
fig = go.Figure(data = data, layout = layout)
iplot(fig)

# Group quiere decir que ponga las barras separadas.
```
